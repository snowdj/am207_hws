### Part A:
Your child has been randomly selected for Type I diabetes screening, using a highly accurate new test that boasts of a false positive rate of 1% and a false negative rate of 0%. The prevalence of of Type I diabetes in children is approximately 0.228%. 

1.Should your child test positive, what is the probability that he/she has Type I diabetes?

2.Should you be concerned enough to ask for further testing or treatment for your child? Suppose an independent test with the same false positive/false negative rate is available.

Later, you read online that Type I diabetes is 6 times more prevalent in prematurely born children. 

3.If this statistic is true, what is the probability that your child, who is prematurely born, has Type I diabetes?

4.Subjectively, given the new information, should you be concerned enough to ask for treatment for your child?

**Justify your decisions using your calculations.**


### Answer to Part A:
Denote the child with or without Type I diabetes as D+ and D- respectively; denote test positive and test negative as T+ and T- respectively; denote prematurely and not prematurely born child as M+ and M- respectively.

We can learn from the problem description that 
$$P(D+) = 0.00228$$
$$P(T+|D-) = 0.01$$
$$P(T-|D+) = 0$$

**1. ** According to Bayes' rule, 

$$P(D+|T+) = \frac{P(T+|D+)P(D+)}{P(T+)} = \frac{P(T+|D+)P(D+)}{P(T+|D-)P(D-)+P(T+|D+)P(D+)}
=\frac{1\cdot 0.00228}{0.01\cdot (1-0.00228) + 1\cdot 0.00228} \approx 0.1860$$

i.e., should the child test ppositive, the probability that the child has Type I diabetes is **0.1860**.


**2. ** 
Intuitively I shouldn't be too concerned since the probability that the child doesn't have Type I diabetes is much higher than otherwise, given positive test result. However, another independent test could be very helpful.

Suppose we perform another independent test, the result could be either positive or negative. We can calculate the probability that the child has Type I diabetes given these results.

If another independent test shows positive result, 

$$P(D+|T+T+) = \frac{P(T+T+|D+)P(D+)}{P(T+T+)} = \frac{P(T+T+|D+)P(D+)}{P(T+T+|D-)P(D-)+P(T+T+|D+)P(D+)}$$
$$=\frac{1^2\cdot 0.00228}{0.01^2\cdot (1-0.00228) + 1^2\cdot 0.00228} \approx 0.9581$$

If another independent test shows negative result,
$$P(D+|T+T-) = \frac{P(T+T-|D+)P(D+)}{P(T+T-)} = \frac{P(T+T-|D+)P(D+)}{P(T+T-|D-)P(D-)+P(T+T-|D+)P(D+)}=0$$

Therefore, I would highly recommend another independent test since we would be much more confident about whether the child has Type I diabetes no matter what result we get.

**3. ** Here we interpret "6 times more prevalent" as "6 times as prevalent". If this statistic is true, we know 
$$P(D+|M+) = 6P(D+) = 0.01368$$

In this case, 

$$P(D+|T+,M+) = \frac{P(T+|D+,M+)P(D+|M+)}{P(T+|M+)} = \frac{P(T+|D+,M+)P(D+|M+)}{P(T+|D-,M+)P(D-|M+)+P(T+|D+,M+)P(D+|M+)}
=\frac{1\cdot 0.01368}{0.01\cdot (1-0.01368) + 1\cdot 0.01368} \approx 0.5811$$

i.e., The probability that the child, who is prematurely born, has Type I diabetes is **0.5811**.

**4. ** I should be concerned enough to ask for treatment since it is more likely that the child has Type I diabetes than not.

### Answer to Part B:

Denote the event that Class A is the class with less harsh graders as S+ (meaning we should Stay), and Class B as S-; denote unfair grade (i.e., lower than the quality of the work) as U+, and fair grade as U-. We know

$$P(U+|S+) = 0.15$$
$$P(U+|S-) = 0.35$$
$$P(S+) = P(S-) = 0.5$$

Since we recieved an unfair grade, we need to compare $P(S+|U+)$ and $P(S-|U+)$. According to Bayes' rule, we have

$$P(S+|U+) = \frac{P(U+|S+)P(S+)}{P(U+|S+)P(S+)+P(U+|S-)P(S-)} = \frac{0.15\cdot0.5}{0.15\cdot0.5+0.35\cdot0.5}=0.3$$

## Problem 2: Maximum Likelihood Estimates

**Coding required for Part B only**

Suppose you observe the following data set $\mathbf{x}^{(0)} = (0.5, 2.5), \mathbf{x}^{(1)} = (3.2, 1.3), \mathbf{x}^{(2)} = (2.72, 5.84), \mathbf{x}^{(3)}= (10.047, 0.354)$. By convention, for any vector $\mathbf{x}$, we will denote the first component of $\mathbf{x}$ by $x_{1}$ and the second component by $x_{2}$. Suppose that the data is drawn from the same two-dimensional probability distribution with pdf $f_X$, that is, $\mathbf{x}^{(i)} \overset{iid}{\sim} f_X$, where
$$
f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\}.
$$
You should assume that $\lambda_1, \lambda_0 > 0$ and that $f_X$ is supported on the nonnegative quandrant of $\mathbb{R}^2$ (i.e. $f_X$ is zero when either component is negative). 

### Part A:

What are the values for $\lambda_0$ and $\lambda_1$ that maximizes the likelihood of the observed data? 

**Support your answer with full and rigorous analytic derivations.**

### Part B:

Visualize the data along with the distribution you determined in part A (in two dimensions or three).


### Answer to Part A:

According to the property of probability density function, we have 
$$\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}{f_X(\mathbf{x})}d x_1 dx_2=1.$$
In this case, 
$$f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} e^{-\lambda_0 (x^2_{1} + x^2_{2})}.$$

$$\therefore \int_{-\infty}^{\infty}\int_{-\infty}^{\infty}{f_X(\mathbf{x})}d x_1 dx_2
=\int_{0}^{\infty}\int_{0}^{\infty}{4\lambda_1^2 x_{1}x_{2} e^{-\lambda_0 (x^2_{1} + x^2_{2})}}d x_1 dx_2
=4\lambda_1^2\int_{0}^{\infty}{x_{1} e^{-\lambda_0 x^2_{1}}dx_1}\int_{0}^{\infty}{x_{2} e^{-\lambda_0 x^2_{2}}dx_2}
=4\lambda_1^2\cdot \frac{1}{2\lambda_0}\cdot \frac{1}{2\lambda_0}=\frac{\lambda_1^2}{\lambda_0^2}=1$$

Considering $\lambda_1, \lambda_0>0$, we get $\lambda_1=\lambda_0$.

$$\therefore\ f_X(\mathbf{x}) =  4\lambda_0^2 x_{1}x_{2} e^{-\lambda_0 (x^2_{1} + x^2_{2})}.$$

We can now write down the log-likelihood function,
$$\ell(\lambda_0) = \sum_{i=1}^n{\ln f_X(\mathbf{x}^{(i)})}
=\sum_{i=1}^n{\ln (4\lambda_0^2 x_{1}^{(i)}x_{2}^{(i)} e^{-\lambda_0 (x^{(i)2}_{1} + x^{(i)2}_{2})})}
=2n\ln \lambda_0 + \sum_{i=1}^n\ln (4x_{1}^{(i)}x_{2}^{(i)}) - \sum_{i=1}^n\lambda_0 (x^{(i)2}_{1} + x^{(i)2}_{2})$$

Let $\frac{d\ell}{d\lambda_0}=0$, we get
$$\frac{d\ell}{d\lambda_0}=\frac{2n}{\lambda_0}-\sum_{i=1}^n(x^{(i)2}_{1} + x^{(i)2}_{2})=0$$

$$\therefore\ \lambda_0 = \lambda_1 = \frac{2n}{\sum_{i=1}^n(x^{(i)2}_{1} + x^{(i)2}_{2})}$$

We can calculate $\lambda_0, \lambda_1$ using our dataset as follows.

```python
x = [(0.5, 2.5), (3.2, 1.3), (2.72, 5.84), (10.047, 0.354)]

def cal_lambda(x):
    '''Calculate MLE of lambda_0
    '''    
    n = len(x)
    s = sum([_x[0]**2+_x[1]**2 for _x in x])
    return 2*n/s

print('lambda_0 = {}'.format(cal_lambda(x)))
```

i.e., when $\lambda_0=\lambda_1=0.049688970337392765$, the likelihood of the observed data is maximized.

### Answer to Part B:

We can generate 10000 bootstrap samples and estimate the mean and the standard deviation of the distribution using these samples.

```python
def get_bootstrap_sample(data):
    '''Returns an individual bootstrap sample
    '''
    return data[np.random.choice(len(data), size=len(data), replace=True)]

def perform_bootstrap_sample(data, M_samples=1000):
    '''Returns M_samples bootstrap samples
    '''
    return np.array([get_bootstrap_sample(data) for _ in range(M_samples)])

M_samples = 10000
data = df.values
np.random.seed(0)
bs = perform_bootstrap_sample(data, M_samples)
mean = np.mean(bs, axis=1)
```

**1. ** We can visualize the mean of each bootstrap sample as follows.

```python
g = sns.jointplot(x=0, y=1, data=pd.DataFrame(mean));
g.ax_joint.collections[0].set_alpha(0.3);
g.set_axis_labels("$\hat{\mu}_{x1}$", "$\hat{\mu}_{x2}$");
```

```python
print('The estimated mean of the distribution f_x1 is {:.4f}.'\
      .format(np.mean(data, axis=0)[0]))
print('The SE of the estimate is {:.4f}.'.format(np.std(mean, axis=0)[0]))
```

**2. ** We can visualize the distribution of the standard deviation of the distribution $f_{x2 \mid x1}$ for $x_1 \in [3.99, 4.01]$ as follows.

```python
# Calculate the standard deviation for each bootstrap sample
std = np.array([np.std(b[(b[:, 0] >= 3.99) & (b[:, 0] <= 4.01), 1]) for b in bs])
```

```python
sns.distplot(std);
plt.xlabel('$\hat{\sigma}_{x_2|x_1}$');
```

```python
print('The estimated standard deviation of the distribution f_x2|x1 for x1 in [3.99, 4.01] \
is \n{:.4f}.'\
      .format(np.std(data[(data[:, 0] >= 3.99) & (data[:, 0] <= 4.01), 1])))
print('The SE of the estimate is {:.4f}.'.format(np.std(std)))
```

**3. ** Given the SE, 1 digit in the standard deviation estimate are significant since the order of magnitude for standard deviation estimate is -1 and the order of magitude for the SE is -2, indicating digits after 2 decimal places are mainly determined by random error and are not reliable.

**Coding required**

Missing data is a very important topic in statistics and machine learning. Let us begin to explore how to handle missing data in our analysis.  We'll be working with a dataset from the UCI Machine Learning Repository that uses a variety of wine chemical predictors to classify wines grown in the same region in Italy.  Each line represents 13 (mostly chemical) predictors of the response variable wine class, including things like alcohol content, hue, and phenols.  Unfortunately some of the predictor values were lost in measurement. Please load [wine_quality_missing.csv](wine_quality_missing.csv).  (If pandas makes your life easier, feel free to use it).

1.  One way to handle missing data is to just totally ignore it by removing any rows that have any missing values.  This is called *drop imputation*.  Use drop imputation on our wine quality dataset.  How many rows does our imputed dataset have?

1. Another way to handle missing data is to replace any missing value with the mean of the non-missing values in that **column**.  This is called *mean imputation*.  How many rows does our mean imputed dataset have?

1.  Empirically estimate the pdf of the distribution of the sample mean of the ash predictor under drop imputation and mean imputation. Use the perform_bootstrap function from problem 3.

### Answer to Problem 4:

We can load data using pandas.

```python
df = pd.read_csv('wine_quality_missing.csv')
df.head()
```

**1. ** We can use drop imputation.

```python
df1 = df.dropna()
print('There are {} rows in the original dataset.'.format(len(df)))
print('After drop imputation, there are {} rows in the imputed dataset.'.format(len(df1)))
```

**2. ** We can use mean imputation.

```python
df2 = df.fillna(df.mean())
print('There are {} rows in the original dataset.'.format(len(df)))
print('After mean imputation, there are {} rows in the imputed dataset.'.format(len(df2)))
```

**3. ** We can generate 2000 bootstrap samples and estimate the pdf of the distribution of the sample mean of the ash predictor under drop imputation and mean imputation using histogram and kernel density estimate.

```python
###########
# (Copied from Problem 3)

def get_bootstrap_sample(data):
    '''Returns an individual bootstrap sample
    '''
    return data[np.random.choice(len(data), size=len(data), replace=True)]

def perform_bootstrap_sample(data, M_samples=1000):
    '''Returns M_samples bootstrap samples
    '''
    return np.array([get_bootstrap_sample(data) for _ in range(M_samples)])

###########

M_samples = 2000
ash1 = df1['Ash'].values
ash2 = df2['Ash'].values

np.random.seed(0)
bs1 = perform_bootstrap_sample(ash1, M_samples)

np.random.seed(0)
bs2 = perform_bootstrap_sample(ash2, M_samples)

mean1 = np.mean(bs1, axis=1)
std1 = np.std(bs1, axis=1)

mean2 = np.mean(bs2, axis=1)
std2 = np.std(bs2, axis=1)
```

```python
plt.figure(figsize=(8, 12))
plt.subplot(2, 1, 1)
sns.distplot(mean1);
plt.xlim(2.2, 2.55);
plt.title('Drop imputation');
plt.subplot(2, 1, 2)
sns.distplot(mean2);
plt.xlim(2.2, 2.55);
plt.title('Mean imputation');
plt.xlabel('$\hat{\mu}$');
```

```python
print('Drop imputation:')
print('The mean of sample mean is {}.'.format(np.mean(mean1)))
print('The SE of sample mean is {}.'.format(np.std(mean1)))
print('The mean of sample standard deviation is {}.'.format(np.mean(std1)))
print('The SE of sample standard deviation is {}.'.format(np.std(std1)))
print()
print('Mean imputation:')
print('The mean of sample mean is {}.'.format(np.mean(mean2)))
print('The SE of sample mean is {}.'.format(np.std(mean2)))
print('The mean of sample standard deviation is {}.'.format(np.mean(std2)))
print('The SE of sample standard deviation is {}.'.format(np.std(std2)))
```

**4. ** We can compare the SE of the sample mean distributions under drop imputation and mean imputation.

```python
print('Drop imputation:')
print('The SE of sample mean is {}.'.format(np.std(mean1)))
print()
print('Mean imputation:')
print('The SE of sample mean is {}.'.format(np.std(mean2)))
```

As we can see, the SE under mean imputation is smaller than that under drop imputation. 

Reasons include:
1. The size of the mean imputed dataset is larger than that of the drop imputed dataset; as a result, the drop imputed dataset is more sensitive to outliers and the SE of sample mean is higher.
2. The mean imputed dataset is dominated by data points with the same value (i.e., the mean value) when we look at a single predictor, making the distribution centralizing towards the mean value, which results in a smaller SE of sample mean.
